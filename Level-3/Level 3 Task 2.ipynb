```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load the dataset

df = pd.read_csv(r"/content/Dataset .csv")
df
```

### Identify the Restaurants with the Highest and Lowest Number of Votes

```python
# Identify the restaurant with the highest and lowest number of votes

highest_votes = df.loc[df["Votes"].idxmax()]
lowest_votes = df.loc[df["Votes"].idxmin()]
```

```python
# Display their details

highest_votes[["Restaurant Name", "Votes", "Aggregate rating"]], lowest_votes[["Restaurant Name", "Votes", "Aggregate rating"]]
```

### Analyze if there is a Correlation Between the Number of Votes and the Rating of a Restaurant

```python
# Check correlation between Votes and Aggregate rating

correlation = df["Votes"].corr(df["Aggregate rating"])
```

```python
# Visualize the relationship using a scatter plot

plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["Votes"], y=df["Aggregate rating"], alpha=0.5)
plt.xlabel("Number of Votes")
plt.ylabel("Aggregate Rating")
plt.title("Correlation Between Votes and Rating")
plt.show()
```
